```python
import matplotlib.pyplot as plt
```

```python
def payoff_plot(strike,premium,s0):
    x=[]
    y=[]

    for i in range(int(s0)-600,int(s0)+600):
        x.append(i)
        if i<=strike:
            y.append(-premium)
        else:
            y.append(i-strike - premium)
    plt.figure(figsize=(16,6))
    plt.plot(x,y)
    
    plt.axhline(0,color="k")
    
    #Premium
    plt.axhline(-premium, ls="dashed", color="blue",lw=1)
    plt.text(int(s0)-400,-premium,-premium,fontsize= 11,verticalalignment="bottom", color="blue")
    
    #Breakeven point
    plt.axvline(strike + premium,ls="dashed",color="r",lw=1)
    plt.text(strike+premium,200, strike+premium, fontsize = 11, verticalalignment="top", color="red")
    
    #Strike Price
    plt.axvline(strike,ls="dashed",color="r",lw=1)
    plt.text(strike,200, strike, fontsize = 11, horizontalalignment="right", color="red")
    
    #Spot Price
    plt.axvline(s0,ls="dashed",color="green",lw=1)
    plt.text(s0,100, s0, fontsize = 11, horizontalalignment="right", color="green")
    
    plt.xlabel("Spot Price at Expiry")
    plt.ylabel("PnL")
    
    if(s0>strike):
        plt.axhline(s0-strike-premium,ls="dashed",color="purple",lw=1)
        plt.text(int(s0)+400,s0-strike-premium,round(s0-strike-premium,2),color="purple", fontsize=10)
        plt.title("Payoff Curve for Call Option, Premium = {}, current PnL = {}".format(premium,round(s0-strike-premium,2)))
    else:
        plt.title("Payoff Curve for Call Option, Premium = {}, current PnL = {}".format(premium,round(-premium,2)))
    plt.tight_layout()
```

```python
s0=14018.50

arr=[[13950,341.8],[14000,304.00],[14050,280.75]]

l=len(arr)
for i in range(l):
    strike=arr[i][0]
    premium = arr[i][1]
    payoff_plot(strike,premium,s0)
```
